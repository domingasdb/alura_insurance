```python
from pandas import DataFrame, read_csv
```

```python
dados = read_csv('clientes.csv', sep=';')
dados
```

```python
dados.shape
```

```python
dados.loc[:,'Sinistros'] = dados['Sinistros'].fillna(dados['Sinistros'].mean()).round()
# dados.loc[:,'Sinistros'] = dados['Sinistros'].fillna(0)
dados
```

```python
dados.loc[:,'Desconto'] = round(dados.Desconto, 4)
dados.loc[:,'Sinistros'] = dados['Sinistros'].round()
dados
```

```python
dados = dados.drop('Desconto', axis=1)
dados
```

```python
# Ordenacao por idades
dados_ordenados = dados.sort_values(['Idade'])
dados_ordenados
```
