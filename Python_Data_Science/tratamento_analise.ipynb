```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
```

```python
dados = read_csv('clientes.csv', sep=';')
dados
```

```python
dados.shape
```

```python
dados['Sinistros'] = dados['Sinistros'].fillna(dados['Sinistros'].mean()).astype(int)
# dados.loc[:,'Sinistros'] = dados['Sinistros'].fillna(0)
dados
```

```python
dados.loc[:,'Desconto'] = round(dados.Desconto, 2)
dados.loc[:,'Sinistros'] = dados['Sinistros'].astype(int)
dados
```

```python
dados = dados.drop('Desconto', axis=1)
dados
```

```python
# Ordenacao por idades
dados_ordenados = dados.sort_values(['Idade'])
dados_ordenados
```

```python
dados.groupby('Estado Civil')['Prêmio'].sum()
```

```python
dados.groupby('Categoria')['Id'].count()
```

```python
plt.scatter(dados['Idade'], dados['Prêmio'])
plt.show()
```

```python
tabela = dados.groupby('Sexo')['Id'].count()
label = tabela.index
valores = tabela.values
plt.pie(valores, labels=label, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```

```python
dados.groupby('Sinistros')['Id'].count()
```

```python
tabela = dados.groupby('Sinistros')['Id'].count()
sinistros = [7,1,2,0,5,6,4,3]
valores = sorted(tabela.values/sum(tabela.values))
plt.bar(sinistros, valores)
plt.show()
```

```python
tabela = dados.groupby('Categoria')['Sinistros'].count()
sinistros = ['Jovem','Adulto','Sênior']
valores = sorted(tabela.values/sum(tabela.values))
plt.bar(sinistros, valores)
plt.show()
```
