```python
from pandas import DataFrame
from numpy import random, arange
```

```python
base = {
    'ids': arange(1, 1001),
    'idades': random.randint(18, 80, size=1000),
    'sexos': random.choice(['M', 'F'], size=1000),
    'estados_civis': random.choice(['Solteiro(a)', 'Casado(a)', 'Solteiro(a)', 'Divorciado(a)', 'Solteiro(a)'],1000),
    'sinistros': random.randint(0, 5, 1000),
    'premios': random.gamma(2, 1000, size=1000)
}
```

```python
base = DataFrame(base)
base.head(10)
```

```python
clientes_maior_30anos = base[base['idades'] > 30]
clientes_maior_30anos
```

```python
clientes_sem_sinistro = base.query('sinistros == 0')
clientes_sem_sinistro
```

```python
for idx, row in base.iterrows():
    print(f"Idade: {row['idades']}, PrÃªmio: {row['premios']}")
```
